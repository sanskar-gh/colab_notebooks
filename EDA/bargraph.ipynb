```python
import matplotlib.pyplot as plt

product = ['A', 'B', 'C', 'D', 'E']
sales = [10000, 20000, 15000, 11000, 23000]
plt.bar(product, sales, color = 'aqua', label = "Sales_2025")
plt.xlabel("Product Name")
plt.ylabel("Monthly Sales")
plt.title("MOnthly Sales for Each Product in 2025")
plt.show()
```
