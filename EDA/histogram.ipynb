```python
import matplotlib.pyplot as plt
scores = [55, 78, 45, 89, 90, 67, 73, 84, 91, 62,
          77, 80, 56, 68, 74, 82, 95, 59, 70, 88]
plt.hist(scores, bins = 5, color = 'red', edgecolor = 'black')
plt.xlabel("Range of scores")
plt.ylabel("Number of students")
plt.title("Scores Vs Students histogram")
plt.show()
```
