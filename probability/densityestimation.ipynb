```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
sample = norm.rvs(loc = 50, scale = 5, size = 1000)
```

```python
s = pd.Series(sample)
```

```python
s.mean()
```

```python
s.plot(kind = 'hist')
```

```python
sample_mean = s.mean()
sample_std = s.std()
```

```python
sample_std
```

```python
sample_mean
```

```python
dist = norm(sample_mean, sample_std)
```

```python
values = np.linspace(sample.min(), sample.max(), 100)
```

```python
sample.max()
```

```python
sample.min()
```

```python
probability_density = [dist.pdf(value) for value in values]
```

```python
probability_density
```

```python
plt.hist(sample, bins = 10, density = True)
plt.plot(values, probability_density)
```
