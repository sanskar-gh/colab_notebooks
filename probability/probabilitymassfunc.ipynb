```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
L = []
for i in range(10000):
    L.append(random.randint(1,6))
len(L)
```

```python
L = pd.Series(L)
```

```python
s = (L.value_counts() / L.value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind = 'bar')
```

# FOR TWO DIE

```python
S = []
for i in range(10000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    S.append(a + b)
len(S)
```

```python
S[:5]
```

```python
S = pd.Series(S)
```

```python
p = (S.value_counts() / S.value_counts().sum()).sort_index()
```

```python
p
```

```python
p.plot(kind = 'bar')
```
